```python
items = df_rd.to_dict(orient='records')
print(len(items))
```

```python
dynamodb = boto3.resource('dynamodb')

table = dynamodb.Table('animals')
```

```python
for item in items:
    table.put_item(Item=item)
```

```python
response = table.scan()
items = response['Items']
len(items)

# AWS sdk will only retrieve 1MB data upon each scan.
# So we will have to perform a thing called pagination to get all the items back.
```

```python
response.keys()
# response['Count']
```

```python
all_items = []

# Perform paginated scan to retrieve all items
response = table.scan()

# Extract items from the first page
all_items.extend(response['Items'])

# Continue paginating through remaining pages
while 'LastEvaluatedKey' in response:
    response = table.scan(ExclusiveStartKey=response['LastEvaluatedKey'])
    all_items.extend(response['Items'])

# Print the total number of items retrieved
print("Total items retrieved:", len(all_items))
```
